## Problem Statement

Given a dataset provided by the Ames Assessor's Office (used by the office to computing assessed values for individual properties sold in Ames, IA from 2006 to 2010), we aim to create a model to best predict actual home sale prices. We will evaluate actual performance against a test dataset where we do not have access to the actual sale price.

## Executive Summary

The data provided by the Ames Assessor's Office for individual properties sold in Ames, IA from 2006-2010 has 2,051 observations and 82 descriptors of the homes, including the final, actual sale price of the individual properties. By using this dataset as both a training and test set, we will use linear regression methods along with feature engineering to best fit a model that predicts prices for another provided dataset that is stripped of the actual sale price data. We will evaluate our model's predicted sale prices versus the actual sale prices via (Root Mean-Squared-Error) and compare our results to others via Kaggle. Simple linear regression models such as these can be useful in pricing analysis on a single-homebuyer level where companies such as Zillow and Trulia can show their estimated housing cost to aid customers with their home transactions. On a government/business level, it can be utilized in terms of taxation or evaluation of growth. In either case, models such as these are valuable if accurate and generalizable.

## Functions

```python
# Obtain all metrics following every test
def eval(y, y_pred):
    print(f"R2 score is: {r2_score(y,y_pred)}")
    print(f"MSE score is: {mean_squared_error(y,y_pred)}")
    print(f"RMSE score is: {np.sqrt(mean_squared_error(y,y_pred))}")
```

```python
# Check check quickly inspect each column to assist in processing
def clean_check(dfs):
    print(dfs.isnull().sum())
    print(dfs.value_counts())
```

```python
# Modify column types slightly easier in case we have to do it multiple times
def change_type(df, column, newtype):
    df[column] = df[column].astype(newtype)
```

```python
# Filter and clean all dataframes
def filter_frame(*dfs):
    for df in dfs: # Iterate through both dataframe objects
        
        # Transform titles, replace spaces with underscores and lowercasing all letters
        df.columns = df.columns.str.replace(' ','_')
        df.columns = df.columns.str.lower()
        
        # Change type for columns
        change_type(df, 'id', str)
        change_type(df, 'pid', str)
        change_type(df, 'ms_subclass', str)
        
        # Fill lot_frontage null values with zeros
        df['lot_frontage'].fillna(0, inplace=True)
        
        # All ordinal values will be mapped from 0-n, where 0 is the least desirable
        # Ordinalize all values that the data dictionary suggested in sources and converted as many to
        # numerical columns as possible to avoid unnecessary dummying
        
        # Fill alley NA's with OTHER and convert to ordinal list 0-2
        df['alley'].fillna("OTHER", inplace=True)
        df['alley'] = df['alley'].map({'Pave': 2, 'Grvl': 1, 'OTHER':0})
        
        # Ordinalize lot shape from 0-3
        df['lot_shape'] = df['lot_shape'].map({'IR3': 0,'IR2': 1, 'IR1':2, 'Reg':3})
        
        # Binarize utilities
        df['utilities'] = df['utilities'].map({'AllPub':1}).fillna(0).astype(int)
        
        # Ordinalize land slope form 0-2
        df['land_slope'] = df['land_slope'].map({'Gtl':2, 'Mod':1, 'Sev':0}).astype(int)
        
        # Overall_cond already ordinal, convert to int for numerical analysis
        df['overall_cond'] = df['overall_cond'].astype(int)
        
        # Ensure year_remod/add is integer for numerical processing
        df['year_remod/add'] = df['year_remod/add'].astype(int)
        
        # Fill roof material with 1's and 0's
        df['roof_matl'] = df['roof_matl'].map({'CompShg':1}).fillna(0).astype(int)
        
        # Convert mas_vnr_type nulls to None.
        df['mas_vnr_type'].fillna("None", inplace=True)
        
        # Fill mas_vnr_area nulls with zeros
        df['mas_vnr_area'].fillna(0, inplace=True)
        
        # Fill exter qual with ordinals (0-4)
        df['exter_qual'] = df['exter_qual'].map({'Ex':4, 'Gd':3, 'TA':2, 'Fa':1, 'Po':0}).astype(int)
        
        # Fill exter cond with ordinals (0-4)
        df['exter_cond'] = df['exter_cond'].map({'Ex':4, 'Gd':3, 'TA':2, 'Fa':1, 'Po':0}).astype(int)
        
        # Convert bsmnt qual NAs to None, indicating no basement and ordinalize 0-5
        df['bsmt_qual'] = df['bsmt_qual'].map({'Ex':5, 'Gd':4, 'TA':3, 'Fa':2, 'Po':1}).fillna(0).astype(int)
        
        # Convert bsmnt cond NAs to None, indicating no basement and ordinalize 0-5
        df['bsmt_cond'] = df['bsmt_cond'].map({'Ex':5, 'Gd':4, 'TA':3, 'Fa':2, 'Po':1}).fillna(0).astype(int)
        
        # Convert bsmt_exposure NAs to None, indicating no basement and ordinalize 0-4
        df['bsmt_exposure'] = df['bsmt_exposure'].map({'Gd':4, 'Av':3, 'Mn':2, 'No':1}).fillna(0).astype(int)
        
        # Convert bsmt_type_1 NAs to None, indicating no basement and ordinalize 0-6
        df['bsmtfin_type_1'] = df['bsmtfin_type_1'].map({'GLQ':6, 'ALQ':5, 'BLQ':4, 'Rec':3, 'LwQ':2, 'Unf':1}).fillna(0).astype(int)
        
        # Fill nulls in bsmtfin_sf_1 with 0
        df['bsmtfin_sf_1'].fillna(0, inplace=True)
        
        # Convert bsmt_type_2 NAs to None, indicating no basement and ordinalize 0-6
        df['bsmtfin_type_2'] = df['bsmtfin_type_2'].map({'GLQ':6, 'ALQ':5, 'BLQ':4, 'Rec':3, 'LwQ':2, 'Unf':1}).fillna(0).astype(int)
        
        # Fill nulls to zero
        df['bsmtfin_sf_2'].fillna(0, inplace=True)
        
        # Fill nulls to zero
        df['bsmt_unf_sf'].fillna(0, inplace=True)
        
        # Fill nulls to zero
        df['total_bsmt_sf'].fillna(0, inplace=True)
        
        # Ordinalize 0-4
        df['heating_qc'] = df['heating_qc'].map({'Ex':4, 'Gd':3, 'TA':2, 'Fa':1, 'Po':0}).astype(int)
        
        # Dummify Central Air manual. 1 for AC, 0 for no AC
        df['central_air'] = df['central_air'].map({'Y': 1}).fillna(0).astype(int)
        
        # Convert one value of electrical to SBrkr and ordinalize 0-4
        df['electrical'] = df['electrical'].fillna("SBrkr")
        df['electrical'] = df['electrical'].map({'SBrkr':4, 'FuseA':3, 'FuseF':2, 'FuseP':1, 'Mix':0})
        
        # Fill null of bsmt_full_bath and conver to int
        df['bsmt_full_bath'] = df['bsmt_full_bath'].fillna(0).astype(int)
        
        # Fill null of bsmt_half_bath and conver to int
        df['bsmt_half_bath'] = df['bsmt_half_bath'].fillna(0).astype(int)
        
        # Convert full_bath to int
        df['full_bath'] = df['full_bath'].astype(int)
        
        # Convert half_ath to int
        df['half_bath'] = df['half_bath'].astype(int)
        
        # Convert bedroom_abvgr to int
        df['bedroom_abvgr'] = df['bedroom_abvgr'].astype(int)
        
        # Convert kitchen_abvgr to int
        df['kitchen_abvgr'] = df['kitchen_abvgr'].astype(int)
        
        # Ordinalize kitchen_qual 0-4
        df['kitchen_qual'] = df['kitchen_qual'].map({'Ex':4, 'Gd':3, 'TA':2, 'Fa':1, 'Po':0}).astype(int)
        
        # Convert to totrms_abvgrd continuous int
        df['totrms_abvgrd'] = df['totrms_abvgrd'].astype(int)
        
        # Ordinalize 0-7 for functional
        df['functional'] = df['functional'].map({'Typ':7, 'Min1':6, 'Min2':5, 'Mod':4, 'Maj1':3, 'Maj2':2, 'Sev':1, 'Sal':0}).astype(int)
        
        # Fill null of fire-lace_qu and ordinalize 0-5
        df['fireplace_qu'] = df['fireplace_qu'].fillna("None")
        df['fireplace_qu'] = df['fireplace_qu'].map({'Ex':5, 'Gd':4, 'TA':3, 'Fa':2, 'Po':1, 'None':0}).astype(int)

        # Convert garage_type NAs to None, indicating no garage
        df['garage_type'] = df['garage_type'].fillna("None")
        
        # Convert garage_yr_blt NAs to 0, indicating no garage. We will probably not use this thing. 
        df['garage_yr_blt'] = df['garage_yr_blt'].fillna(0).astype(int)
       
        # Convert garage_finish NAs to None, indicating no garage ordinalize 0-3
        df['garage_finish'] = df['garage_finish'].fillna("None")
        df['garage_finish'] = df['garage_finish'].map({'Fin':3, 'RFn':2, 'Unf':1, 'None':0}).astype(int)
        
        # Fill nulls in garage cars with zero
        df['garage_cars'].fillna(0, inplace=True)
        
        # Fill nulls in garage area with zero
        df['garage_area'].fillna(0, inplace=True)
        
        # Fill garage_qual with none and ordinalize 0-5
        df['garage_qual'] = df['garage_qual'].fillna("None")
        df['garage_qual'] = df['garage_qual'].map({'Ex':5, 'Gd':4, 'TA':3, 'Fa':2, 'Po':1, 'None':0}).astype(int)
        
        # Fill garage_cond with none and ordinalize 0-5
        df['garage_cond'] = df['garage_cond'].fillna("None")
        df['garage_cond'] = df['garage_cond'].map({'Ex':5, 'Gd':4, 'TA':3, 'Fa':2, 'Po':1, 'None':0}).astype(int)

        # Ordinalize paved_dave 0-2
        df['paved_drive'] = df['paved_drive'].map({'Y':2, 'P':1, 'N':0}).astype(int)
        
        # Fill nulls with none and ordinalize all with 0-4
        df['pool_qc'] = df['pool_qc'].fillna("None")
        df['pool_qc'] = df['pool_qc'].map({'Ex':4, 'Gd':3, 'TA':2, 'Fa':1, 'None':0}).astype(int)
        
        # Fill nulls with none and ordianlize all with 0-4
        df['fence'] = df['fence'].fillna("None")
        df['fence'] = df['fence'].map({'GdPrv':4, 'MnPrv':3, 'GdWo':2, 'MnWw':1, 'None':0}).astype(int)
        
        # Fill nulls with none
        df['misc_feature'].fillna("None", inplace=True)
        
        # Convert to categorical (because this dataset covers multiple years, 
        # mo sold as a continous does not make sense)
        df['mo_sold'] = df['mo_sold'].astype(str)
        
        # Convert year sold to int, as this suggest the age of the house.
        df['yr_sold'] = df['yr_sold'].astype(int)
        
        # Drop id and pid, since these are just labels. Avoid dummying them or having them in analysis.
        df.drop(columns=['id', 'pid'],inplace=True, axis=1)

    #   Optional drop commands for testing.    
    #   df.drop(columns=['id', 'pid', 'street', 'lot_shape', 'land_contour', 
    #                    'utilities', 'land_slope', 'year_remod/add', 
    #                    'foundation', 'pool_area', 'pool_qc', 'garage_yr_blt'], inplace=True, axis=1)
```

```python
# Assemble final csv for submission to kaggle.
def assemble_answer(final_answer):
    sample_sub['SalePrice'] = pd.Series(final_answer)
    sample_sub.to_csv('./datasets/final_answer.csv', index=False)
```

### Match Dataframes

```python
# Matches columns between two dataframes so that linear regression can run
def match_dataframes(df1, df2):
    # Match dataframe one to dataframe two
    missing_cols = set(list(df1.columns)) - set(list(df2.columns))
    for i in list(missing_cols):
        df2[i] = 0
    
    # Match dataframe two to dataframe one
    missing_cols = set(list(df2.columns)) - set(list(df1.columns))
    for i in list(missing_cols):
        df1[i] = 0    
```

## Single Linear Regression

Simple analysis to get a baseline regression to compare further MLR and regularization methods. Also used to test single predictors individually.

```python
X = X_train
y = y_train_pt
Xt = X_test
yt = y_test_pt
```

```python
lr = LinearRegression()

lr.fit(X, y)

linear_train_preds = lr.predict(X)
linear_test_preds = lr.predict(Xt)
linear_final_preds = lr.predict(X_test)

eval(y_train, pt2_y.inverse_transform(linear_train_preds.reshape(-1,1)))
print("")
eval(y_test, pt2_y.inverse_transform(linear_test_preds.reshape(-1,1)))

# print(lr.score(X_test[features], y_test))
print("")
print(f"Cross_val score: {np.mean(cross_val_score(estimator=lr, X=X, y=y, cv=5))}")

assemble_answer(pt2_y.inverse_transform(linear_final_preds.reshape(-1,1)).flatten())
```

## Multiple Linear Regression

```python
X = X_train_MLR
y = y_train_pt_MLR
Xt = X_test_MLR
yt = y_test_pt_MLR # Lasso
```

```python
# Cross-validate over our list of Lasso alphas.
lasso_model = LassoCV(n_alphas=200, cv=5)

# lasso_model = GridSearch(lasso, 
# Fit model using best ridge alpha!
lasso_model.fit(X, y[:, 0])

lasso_train_preds = lasso_model.predict(X)
lasso_test_preds = lasso_model.predict(Xt)
lasso_final_preds = lasso_model.predict(X_test_MLR)

# Evaluate model.
eval(y_train, pt_y.inverse_transform(lasso_train_preds.reshape(-1,1)))
print("")
eval(y_test, pt_y.inverse_transform(lasso_test_preds.reshape(-1,1)))
print("")
print(f"Cross_val score: {np.mean(cross_val_score(estimator=lasso_model, X=X, y=y[:, 0], cv=5))}")

assemble_answer(pt_y.inverse_transform(lasso_final_preds.reshape(-1,1)).flatten())
```

```python
#Absolute value and count how many predictors remain after function is regularized.
count = 0
for i in range(len(lasso_model.coef_)):
    if np.absolute(lasso_model.coef_[i])>0:
        count += 1
print(count)
```

```python
# Plot the model
plt.figure(figsize=(6,6))

# Generate a scatterplot of predicted values versus actual values.
plt.scatter(y_test, pt_y.inverse_transform(lasso_test_preds.reshape(-1,1)), s=30, c='r', marker='+', zorder=10, alpha = 0.2)
plt.xlabel("True")
plt.ylabel("Predicted")
plt.title("Model Predictions vs True Value")

plt.plot([0, np.max(y)], [0, np.max(y)], c = 'k')

plt.show();
```

# Ridge

```python
X = X_train
y = y_train_pt
Xt = X_test
yt = y_test_pt
```

```python
# Set up a list of ridge alphas to check.
r_alphas = np.logspace(0, 5, 100)
# Generates 100 values equally between 0 and 5,
# then converts them to alphas between 10^0 and 10^5.

# Cross-validate over our list of ridge alphas.
ridge_model = RidgeCV(alphas=r_alphas, scoring='neg_mean_squared_error', cv=5)

# # Fit model using best ridge alpha!
ridge_model.fit(X,y)
ridge_optimal_alpha = ridge_model.alpha_

# # Instantiate.
ridge_model = Ridge(alpha=ridge_optimal_alpha)

# Fit.
ridge_model.fit(X, y)

# Generate predictions
ridge_train_preds = ridge_model.predict(X)
ridge_test_preds = ridge_model.predict(Xt)
ridge_final_preds = ridge_model.predict(test)

# Evaluate model using R
eval(y_train, pt_y.inverse_transform(ridge_train_preds.reshape(-1,1)))
print("")
eval(y_test, pt_y.inverse_transform(ridge_test_preds.reshape(-1,1)))
# print(lasso_model.score(X_test, yt))
print("")

print(np.sqrt(-np.mean(cross_val_score(estimator=ridge_model, 
                        X=X, 
                        y=y, 
                        scoring='neg_mean_squared_error',
                        cv=5))))

assemble_answer(pt_y.inverse_transform(ridge_final_preds.reshape(-1,1)).flatten())
```
